```python
def training_evaluation(actual, predicted):

 mse = mean_squared_error(actual, predicted)
 rmse = np.sqrt(mean_squared_error(actual, predicted))
 mae = mean_absolute_error(actual, predicted)
 r2 = r2_score(actual, predicted)

 print(f'MSE: {mse}')
 print(f'RMSE: {rmse}')
 print(f'MAE: {mae}')
 print(f'Training R2 Accuracy: {round(r2_score(actual, predicted), 4)*100} %')
```

```python
training_evaluation(y_train, y_train_pred)
```

```python
print(f'Test Accuracy: {round(r2_score(y_test, y_test_pred), 2)*100} %')
```

```python
def test_evaluation(actual, predicted):

 mse = mean_squared_error(actual, predicted)
 rmse = np.sqrt(mean_squared_error(actual, predicted))
 mae = mean_absolute_error(actual, predicted)
 r2 = r2_score(actual, predicted)

 print(f'MSE: {mse}')
 print(f'RMSE: {rmse}')
 print(f'MAE: {mae}')
 print(f'Testing R2 Accuracy: {round(r2_score(actual, predicted), 2)*100} %')
```
